```python
from sentence_transformers import SentenceTransformer, util

# Cargar modelo preentrenado multilingüe (funciona para muchos idiomas, incluido el quechua)
model = SentenceTransformer('distiluse-base-multilingual-cased-v1')

# Lista de oraciones en español y quechua
spanish_sentences = [
    "El gato está en la casa.",
    "Estoy estudiando.",
    "El sol brilla en el cielo."
]

quechua_sentences = [
    "Misi wasipi kachkan.",  # El gato está en la casa.
    "Yachaykuyta ruwanqani.",  # Estoy estudiando.
    "Inti hanaqpachapi k'anchan."  # El sol brilla en el cielo.
]

# Generar embeddings para las oraciones en español y quechua
spanish_embeddings = model.encode(spanish_sentences, convert_to_tensor=True)
quechua_embeddings = model.encode(quechua_sentences, convert_to_tensor=True)

# Calcular la similitud de coseno entre cada oración en español y quechua
cosine_scores = util.pytorch_cos_sim(spanish_embeddings, quechua_embeddings)

# Mostrar las puntuaciones de similitud (matriz de similitud)
print("Matriz de Cosine Similarity:")
print(cosine_scores)
```

```python
from transformers import M2M100ForConditionalGeneration, M2M100Tokenizer
import torch
from sentence_transformers import SentenceTransformer, util

# Cargar el modelo M2M-100 y el tokenizer
model_name = "facebook/m2m100_418M"
tokenizer = M2M100Tokenizer.from_pretrained(model_name)
model = M2M100ForConditionalGeneration.from_pretrained(model_name)

# Lista de oraciones en español
spanish_sentences = [
    "El gato está en la casa.",
    "Estoy estudiando.",
    "El sol brilla en el cielo."
]

# Función para traducir oraciones usando M2M-100
def translate_m2m(sentences, source_lang="es", target_lang="en"):
    translations = []
    tokenizer.src_lang = source_lang  # Configurar idioma de origen (español)

    for sentence in sentences:
        # Tokenizar la oración y generar la traducción
        encoded = tokenizer(sentence, return_tensors="pt")
        generated_tokens = model.generate(**encoded, forced_bos_token_id=tokenizer.get_lang_id(target_lang))  # Configurar idioma objetivo (inglés)

        # Decodificar las traducciones a texto
        translated_text = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
        translations.append(translated_text[0])

    return translations

# Traducir oraciones del español al inglés usando M2M-100
translated_sentences_to_en = translate_m2m(spanish_sentences, source_lang="es", target_lang="en")

# Aquí deberías agregar el paso para traducir del inglés al quechua si tienes un modelo que lo haga
# Suponiendo que tienes un modelo que traduce del inglés al quechua, llamarlo aquí.
#translated_sentences_to_qu = translate_m2m(translated_sentences_to_en, source_lang="en", target_lang="qu")

# Mostrar las traducciones al inglés
print("Traducciones al inglés:")
print(translated_sentences_to_en)
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch
from sklearn.metrics.pairwise import cosine_similarity

# Cargar el modelo mT5 pequeño y el tokenizador
model_name = "google/mt5-small"  # Puedes probar con versiones más grandes como "mt5-base"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Lista de oraciones en español y quechua
spanish_sentences = [
    "El gato está en la casa.",
    "Estoy estudiando.",
    "El sol brilla en el cielo."
]

quechua_sentences = [
    "Misi wasipi kachkan.",  # El gato está en la casa.
    "Yachaykuyta ruwanqani.",  # Estoy estudiando.
    "Inti hanaqpachapi k'anchan."  # El sol brilla en el cielo.
]

# Función para generar embeddings usando mT5
def generate_embeddings(sentences):
    embeddings = []
    for sentence in sentences:
        # Tokenizar la oración y generar las representaciones (embeddings)
        input_ids = tokenizer(sentence, return_tensors="pt").input_ids
        with torch.no_grad():
            # Utiliza la capa de embeddings del modelo
            outputs = model.get_input_embeddings()(input_ids).mean(dim=1)  # Promedia sobre la longitud de la secuencia
            embeddings.append(outputs)

    return torch.cat(embeddings, dim=0)

# Generar embeddings para las oraciones en español y quechua
spanish_embeddings = generate_embeddings(spanish_sentences)
quechua_embeddings = generate_embeddings(quechua_sentences)

# Calcular la similitud del coseno entre cada oración en español y quechua
cosine_scores = cosine_similarity(spanish_embeddings.numpy(), quechua_embeddings.numpy())

# Mostrar la matriz de similitud
print("Matriz de Cosine Similarity:")
print(cosine_scores)
```
